```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm
import sympy as sp

class SystemeEtat:
    def __init__(self, A, B, C, D=None):
        self.A = np.array(A)
        self.B = np.array(B)
        self.C = np.array(C)
        self.n = self.A.shape[0]
        self.D = np.array(D) if D is not None else np.zeros((self.C.shape[0], self.B.shape[1]))

    def matrice_transition_scipy(self, t):
        return expm(self.A * t)

    def matrice_transition_cayley_hamilton(self, t):
        s = sp.Symbol('s')
        I = sp.eye(self.n)
        A_sym = sp.Matrix(self.A)
        poly_char = (s * I - A_sym).det()
        coeffs = sp.Poly(poly_char, s).all_coeffs()
        coeffs = [float(c) for c in coeffs]

        eigenvals = np.linalg.eigvals(self.A)
        if self.n == 2:
            lambda1, lambda2 = eigenvals
            exp_lambda1_t = np.exp(lambda1 * t)
            exp_lambda2_t = np.exp(lambda2 * t)
            matrix_sys = np.array([[1, lambda1], [1, lambda2]])
            rhs = np.array([exp_lambda1_t, exp_lambda2_t])
            alphas = np.linalg.solve(matrix_sys, rhs)
            return alphas[0] * np.eye(self.n) + alphas[1] * self.A
        else:
            return self.matrice_transition_scipy(t)

    def matrice_transition_laplace(self, t):
        s, tau = sp.symbols('s t')
        A_sym = sp.Matrix(self.A)
        I = sp.eye(self.n)
        resolvante = (s * I - A_sym).inv()
        phi_sym = resolvante.applyfunc(lambda f: sp.inverse_laplace_transform(f, s, tau))
        phi_num = np.array(phi_sym.subs(tau, t)).astype(np.float64)
        return phi_num

    def reponse_libre(self, x0, t_array, methode='scipy'):
        x0 = np.array(x0).reshape(-1, 1)
        x_response = np.zeros((len(t_array), self.n))

        for i, t in enumerate(t_array):
            if methode == 'scipy':
                phi_t = self.matrice_transition_scipy(t)
            elif methode == 'laplace':
                phi_t = self.matrice_transition_laplace(t)
            else:
                phi_t = self.matrice_transition_cayley_hamilton(t)
            x_t = phi_t @ x0
            x_response[i, :] = x_t.flatten()

        return x_response

    def afficher_comparaison(self, t_max=5, nb_points=100):
        t = np.linspace(0, t_max, nb_points)
        x0 = [1, 0]

        x_scipy = self.reponse_libre(x0, t, 'scipy')
        x_cayley = self.reponse_libre(x0, t, 'cayley')
        x_laplace = self.reponse_libre(x0, t, 'laplace')

        plt.figure(figsize=(14, 10))

        plt.subplot(2, 2, 1)
        plt.plot(t, x_scipy[:, 0], 'b-', label='Scipy', linewidth=2)
        plt.plot(t, x_cayley[:, 0], 'r--', label='Cayley-Hamilton', linewidth=2)
        plt.plot(t, x_laplace[:, 0], 'g-.', label='Laplace', linewidth=2)
        plt.title('État x₁(t)')
        plt.xlabel('Temps (s)')
        plt.ylabel('x₁')
        plt.legend()
        plt.grid(True)

        plt.subplot(2, 2, 2)
        plt.plot(t, x_scipy[:, 1], 'b-', label='Scipy', linewidth=2)
        plt.plot(t, x_cayley[:, 1], 'r--', label='Cayley-Hamilton', linewidth=2)
        plt.plot(t, x_laplace[:, 1], 'g-.', label='Laplace', linewidth=2)
        plt.title('État x₂(t)')
        plt.xlabel('Temps (s)')
        plt.ylabel('x₂')
        plt.legend()
        plt.grid(True)

        plt.subplot(2, 2, 3)
        err_cayley = np.abs(x_scipy - x_cayley)
        err_laplace = np.abs(x_scipy - x_laplace)
        plt.semilogy(t, err_cayley[:, 0], 'r-', label='Erreur x₁ (Cayley)')
        plt.semilogy(t, err_laplace[:, 0], 'g-', label='Erreur x₁ (Laplace)')
        plt.semilogy(t, err_cayley[:, 1], 'r--', label='Erreur x₂ (Cayley)')
        plt.semilogy(t, err_laplace[:, 1], 'g--', label='Erreur x₂ (Laplace)')
        plt.title('Erreurs absolues')
        plt.xlabel('Temps (s)')
        plt.ylabel('Erreur')
        plt.legend()
        plt.grid(True)

        plt.subplot(2, 2, 4)
        plt.plot(x_scipy[:, 0], x_scipy[:, 1], 'b-', linewidth=2, label='Trajectoire')
        plt.plot(x0[0], x0[1], 'ro', markersize=8, label='Condition initiale')
        plt.plot(0, 0, 'ks', markersize=8, label='Point d\'équilibre')
        plt.title('Portrait de phase')
        plt.xlabel('x₁')
        plt.ylabel('x₂')
        plt.legend()
        plt.grid(True)

        plt.tight_layout()
        plt.show()


# === APPLICATION ===
if __name__ == "__main__":
    A = [[0, 1], [-2, -3]]
    B = [[0], [1]]
    C = [[1, 0]]
    sys = SystemeEtat(A, B, C)

    print("=== SYSTÈME D'ÉTAT ===")
    print(f"Matrice A:\n{sys.A}")
    print(f"Matrice B:\n{sys.B}")
    print(f"Matrice C:\n{sys.C}")

    eigenvals = np.linalg.eigvals(sys.A)
    print(f"\nValeurs propres: {eigenvals}")
    print(f"Système stable: {all(np.real(eigenvals) < 0)}")

    t_test = 1.0
    phi_scipy = sys.matrice_transition_scipy(t_test)
    phi_cayley = sys.matrice_transition_cayley_hamilton(t_test)
    phi_laplace = sys.matrice_transition_laplace(t_test)

    print(f"\n=== MATRICE DE TRANSITION À t = {t_test}s ===")
    print("Méthode Scipy:")
    print(phi_scipy)
    print("\nMéthode Cayley-Hamilton:")
    print(phi_cayley)
    print("\nMéthode Laplace:")
    print(phi_laplace)

    err_cayley = np.max(np.abs(phi_scipy - phi_cayley))
    err_laplace = np.max(np.abs(phi_scipy - phi_laplace))
    print(f"\nErreur max Cayley: {err_cayley:.2e}")
    print(f"Erreur max Laplace: {err_laplace:.2e}")

    sys.afficher_comparaison()

    tau = -1 / np.real(eigenvals)
    print(f"\n=== ANALYSE TEMPORELLE ===")
    print(f"Constantes de temps: {tau} secondes")
    print(f"Temps de réponse (95%): {3 * np.max(tau):.2f} secondes")
```
